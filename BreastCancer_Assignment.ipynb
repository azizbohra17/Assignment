```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
```

```python
data_frame = pd.read_csv("breastcancer.csv")
```

```python
data_frame.head(15)
```

```python
data_frame = data_frame.drop('id',1)
data_op = data_frame['class']
data_ip = data_frame.drop('class',1)
data_ip.replace('?',-99999,inplace=True)
print(data_op.head(8))
data_ip.head(10)
```

```python
op = []
for out in data_op:
    if out == 2:
        op.append([1, 0])
    elif out == 4:
        op.append([0, 1])
```

```python
feature = np.array(data_ip)
label = np.array(op)
print(feature.shape,label.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(feature, label, test_size=0.3, random_state=17)
```

```python
x_train.shape
```

```python
design = Sequential()
design.add(Dense(500, input_dim =9, activation='sigmoid'))
design.add(Dense(100,activation='tanh'))
design.add(Dense(2,activation='softmax'))
design.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
design.fit(feature, label, epochs=1200, batch_size=75, validation_data=(x_test, y_test))
```

```python
examine_try1 = np.array([[5,1,1,1,2,1,3,1,1]])
examine_try3 = np.array([[3,1,1,1,2,2,3,1,1]])
examine_try5 = np.array([[4,1,1,3,2,1,3,1,1]])
examine_try7 = np.array([[1,1,1,1,2,10,3,1,1]])

result =design.predict_classes(examine_try5)
```

```python
if result == 0:
    print("Benign")
else :
    print("Malignant")
```
